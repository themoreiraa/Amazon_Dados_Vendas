# 2 ETAPA - TRANSFORMAÇÃO DOS DADOS 

```python
# transformando a minha coluna data para o tipo correto (data)
df["Order Date"] = pd.to_datetime(df["Order Date"], dayfirst=False)

# transformando a minha coluna data para o tipo correto (data)
df["Ship Date"] = pd.to_datetime(df["Ship Date"], dayfirst=False)
```

```python
# VERIFICANDO SE OS MEUS DADOS FORAM CONVERTIDO PARA O TIPO DATA 
df.info()
```

```python
# lendo o meu conjunto de dados 
df.head()
```

```python
# renomeando os nomes das minhas colunas 
df.rename(columns={
    "Region": "Regiao",
    "Country": "Pais",
    "Item Type": "Tipo_Produto",
    "Sales Channel": "Canal_Vendas",
    "Order Priority": "Prioridade_Pedido",
    "Order Date": "Data_Pedido",
    "Order ID": "ID_Pedido",
    "Ship Date": "Data_Envio",
    "Units Sold": "Unidades_Vendidas",
    "Unit Price": "Preco_Unitario",
    "Unit Cost": "Custo_Unitario",
    "Total Revenue": "Receita_Total",
    "Total Cost": "Custo_Total",
    "Total Profit": "Lucro_Total"
}, inplace=True)
```

```python
# Removendo a minha coluna id_pedido
df = df.drop("ID_Pedido", axis= 1)
```

# 3 ETAPA - EXPORTAÇÃO DOS MEUS DADOS LIMPO

```python
# salvando o meu dataset em um arquivo csv 
df.to_csv("dados_tratados.csv",index=False)
```
